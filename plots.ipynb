```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_nodes = [5, 10]
capacity = [5, 10, 20]
chain_length = [101, 51, 24, 82, 42, 21]
trans_throughputs = [98.6, 99.162, 91.416, 141.58, 140.3, 136.302]
b_throughputs = [19.9, 10.116, 4.386, 11.49, 5.92, 2.882]
```

```python
#for number of nodes 5 and 10 respectively, plot throughput for each capacity
#Transactions

# Define bar width
bar_width = 0.2

# Define x positions for the bars
x = np.arange(len(n_nodes))

fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for each capacity
for i, cap in enumerate(capacity):
    ax.bar(x + i * bar_width, trans_throughputs[i::len(capacity)], bar_width, label=f'Capacity {cap}')

# Add labels, title, and legend
ax.set_xlabel('Number of Nodes')
ax.set_ylabel('Throughput')
ax.set_title('Transaction Throughput by Number of Nodes and Capacity')
ax.set_xticks(x + (len(capacity) - 1) * bar_width / 2)
ax.set_xticklabels(n_nodes)
ax.legend()

ax.set_ylim(0, 150)
ax.set_yticks(np.arange(0, 151, 10))

# Show plot
plt.show()
```

```python
#Blocks

# Define bar width
bar_width = 0.2

# Define x positions for the bars
x = np.arange(len(n_nodes))

fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for each capacity
for i, cap in enumerate(capacity):
    ax.bar(x + i * bar_width, b_throughputs[i::len(capacity)], bar_width, label=f'Capacity {cap}')

# Add labels, title, and legend
ax.set_xlabel('Number of Nodes')
ax.set_ylabel('Throughput')
ax.set_title('Block Throughput by Number of Nodes and Capacity')
ax.set_xticks(x + (len(capacity) - 1) * bar_width / 2)
ax.set_xticklabels(n_nodes)
ax.legend()

ax.set_ylim(0, 25)
# Show plot
plt.show()
```
